```python
import pandas as pd
```

```python
df = pd.read_csv("DataSets/Movies.csv")
df.drop_duplicates(inplace=True)
df["YEAR"] = df["YEAR"].str.extract(r'(\d+)').astype(float)
df["ONE-LINE"] = df["ONE-LINE"].str.replace('\n', '')
df["STARS"] = df["STARS"].str.replace('\n', '').str.replace('|', '')
df["STARS"] = df["STARS"].str.extract(r'Director:(.*?)Stars')
df["GENRE"] = df["GENRE"].str.replace('\n', '').str.strip()
df["GENRE"] = df["GENRE"].fillna("Unknown")
df["VOTES"] = df["VOTES"].str.replace(',', '').astype(float)
df["VOTES"] = df["VOTES"].fillna(df["VOTES"].median()).astype(float)
df = df.drop(columns=["Gross"])
df['RATING'] = df['RATING'].fillna(df['RATING'].mean()).astype(float)
df['RunTime'] = df['RunTime'].fillna(df['RunTime'].median()).astype(int)
df['YEAR'] = df['YEAR'].fillna(0)
df['STARS'] = df['STARS'].fillna(0)
df.info()
```
