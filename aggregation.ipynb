# Aggreation 

```python
import pandas as pd
```

```python
df = pd.read_csv("DataSets/Employee.csv")
df
```

```python
g = df.groupby("City")
```

```python
df.groupby("City")["Education"].count()
```

```python
g =  df.groupby("City")["Education"].size()
```

```python
print(df.groupby("City")["Education"].max())
print(df.groupby("City")["Education"].min())
print(df.groupby("City")["Education"].sum())
```

```python
df.groupby("Education")["Age"].agg(['min','max',"mean","median"])
```

```python
df.groupby("City").agg({
    "Age":"max",
    "Education":"min",
    "PaymentTier":"mean"
})
```

```python
df.groupby("Education").agg(
    min_age = ('Age',"min"),
    max_payment = ('PaymentTier',"max"),
    avg_payment = ('PaymentTier',"mean")
)
```

```python
df.groupby("City")["PaymentTier"].mean().reset_index()
```

```python
df.groupby("City")["PaymentTier"].mean().reset_index(name="Avg_PaymentTier")
```
