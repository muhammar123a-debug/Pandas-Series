Filtering and Boolean Indexing

```python
import pandas as pd
```

```python
data = {
    "Name":["Ammar","Kashif","furqan","Arif"],
    "Age":[25, 32, 22, 28],
    "Salary":[40000, 50000, 60000, 70000],
    "Department":["IT","HR","Finance","IT"]
}
df = pd.DataFrame(data)
```

```python
#Filter rows where Salary > 50000
salary_con = df[df["Salary"] > 50000]
# Filter rows where Department is IT
dep_con = df[df["Department"] == "IT"]
# Filter rows where Age < 28
age_con = df[df["Age"] < 28]
# Filter specific column after filtering
sp_col_con = df[df["Salary"] > 50000]["Department"]
print(sp_col_con)
```

BOOLEAN INDEXING (Pandas)

Multiple conditions (AND)

```python
multi = df[(df["Department"] == "IT") & (df["Age"] > 20) & (df["Salary"] > 40000)]
print(multi)
```

Multiple conditions (OR)

```python
df[(df["Department"] == "AI")|(df["Age"] > 20)]
```

NOT condition

```python
df[~(df["Age"] > 30)&(df["Department"] == "IT")]
```

Boolean indexing with multiple columns selection

```python
df.loc[df["Salary"] > 50000, ["Name","Salary","Department"]]
```
