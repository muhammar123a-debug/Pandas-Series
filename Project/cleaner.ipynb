```python
import pandas as pd
```

```python
df = pd.read_csv("../DataSets/Dirty.csv")

print("Original Shape",df.shape)

#clean column
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("[^a-z0-9_]","",regex=True)
)
print("Column Name Cleaned")
df.drop_duplicates(inplace=True)
print("remove Duplicate")
text_cols = df.select_dtypes(include="object").columns
num_cols = df.select_dtypes(include=["int64","float64"]).columns
print("Text column",text_cols)
print("numeric column",num_cols)
#text cleaning automatice
for col in text_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.lower()
        .str.replace("[^a-z0-9@._ ]","", regex=True)
        .str.replace("nan","")
    )

print("Column Clean")
#numeric convertion
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="ignore")
print("Numeric casting attempted")

for col in df.select_dtypes(["int64","float"]).columns:
    if df[col].notna().sum() > 0:
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col]
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].fillna("Unknown")

print("missing value handle")

for col in df.columns:
    try:
        df[col] = pd.to_datetime(col)
    except:
        pass
print("Done converstion checked")

#create clean file 
df.to_csv("Cleaned.csv", index=False)
print("Data cleaned and save Cleaned.csv file")
print("Final Shape", df.shape)
```
