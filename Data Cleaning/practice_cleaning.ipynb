Data Cleaning Practice

```python
import pandas as pd
```

```python
df = pd.read_csv("../DataSets/Dirty.csv")

df.duplicated().sum()
df.drop_duplicates()
df = df.rename(columns={"All Time Peak":"Time_Peak"})
df.info()
```

```python
df["Time_Peak"] = df["Time_Peak"].str.extract("(\d+)")
df["Time_Peak"] = df["Time_Peak"].fillna(0).astype(int)
```

```python
df["Peak"] = df["Peak"].str.extract("(\d+)")
df["Peak"]  = df["Peak"].fillna(0).astype(int)
```

```python
df.columns.str.strip()
df.columns.tolist()
```

```python
df = df.rename(columns={"Actual\xa0gross":"Actual_gross"})
df = df.rename(columns={"Adjusted\xa0gross (in 2022 dollars)":"Adjusted_gross"})
```

```python
df.head()
```

```python
df["Actual_gross"].str.replace("$","").str.replace(",","")
```

```python
df["Actual_gross"] = df["Actual_gross"].str.replace("[^0-9]","", regex=True)
df['Actual_gross'] = pd.to_numeric(df['Actual_gross'], errors='coerce')
```
