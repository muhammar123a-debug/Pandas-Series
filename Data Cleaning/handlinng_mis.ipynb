```python
import pandas as pd
```

```python
df = pd.read_csv("../DataSets/Dirty.csv")
df.head()
```

```python
df["Peak"] = df["Peak"].str.extract("(\d+)")
df["Peak"] = df["Peak"].fillna(0)
```

```python
df.columns = df.columns.str.strip()
print(df.columns.tolist())
```

```python
df["Actual\xa0gross"] = df["Actual\xa0gross"].str.replace("$","").str.replace(",","")

df["Actual\xa0gross"]
```

```python
df["Actual\xa0gross"] = pd.to_numeric(df["Actual\xa0gross"], errors='coerce')
```

```python
df["Actual\xa0gross"] = df["Actual\xa0gross"].fillna(df["Actual\xa0gross"].median())
df["Actual\xa0gross"].isnull()
```

```python
df["All Time Peak"] = df["All Time Peak"].str.extract("(\d+)")
df["All Time Peak"] = df["All Time Peak"].fillna(0)
df.info()
```

```python
df["Adjusted\xa0gross (in 2022 dollars)"] = df["Adjusted\xa0gross (in 2022 dollars)"].str.replace("$","").str.replace(",","")
df["Adjusted\xa0gross (in 2022 dollars)"].astype("float")
```

```python
df["Average gross"] = df["Average gross"].str.replace("$","").str.replace(",","")
df["Average gross"].astype("float")
```

```python
df.isnull().sum()
```
