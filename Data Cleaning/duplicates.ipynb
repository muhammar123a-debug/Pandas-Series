Handling Duplicates (Pandas)

```python
import pandas as pd
```

```python
df = pd.read_csv("../DataSets/Dirty.csv")
df.head(3)
```

```python
df.columns.str.strip()
df.columns.tolist()
```

```python
df["All Time Peak"] = df["All Time Peak"].str.extract("(\d+)")
df["All Time Peak"] = df["All Time Peak"].fillna(0).astype('int64')
```

```python
df["Actual\xa0gross"] = df["Actual\xa0gross"].str.replace("$", "").str.replace(",","")
```

```python
df["Actual\xa0gross"] = pd.to_numeric(df["Actual\xa0gross"], errors="coerce")
df["Actual\xa0gross"].fillna(df["Actual\xa0gross"].median())
```

```python
df["Peak"] = df["Peak"].str.extract("(\d+)")
df["Peak"] = df["Peak"].fillna(0).astype("int64")

```

```python
df = df.rename(columns={'Adjusted\xa0gross (in 2022 dollars)':'num_gross'})
df = df.rename(columns={'All Time Peak':'num_peak'})
df = df.rename(columns={"Year(s)":"year's"})

```
