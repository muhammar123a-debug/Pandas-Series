Data Cleaning Practice

```python
import pandas as pd
```

1️⃣ Handling Missing Values

```python
data = pd.read_csv("../DataSets/Titanic-Dataset.csv")
df = pd.DataFrame(data)
df.head()
df.info()
```

```python
# missing values or duplicates check
#missing value check
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median()).astype("int64")
```

```python
df["Cabin"] = df["Cabin"].fillna("Unkown")
df.isnull().sum()
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df.isnull().sum()
```

Remove Duplicate values

```python
df.duplicated().sum()
```

Data type connverstion

```python
df["Pclass"] = df["Pclass"].astype("category")
df["Sex"] = df["Sex"].astype("category")
df["Embarked"] = df["Embarked"].astype("category")
```

```python
df.dtypes
```

String/Text Cleaning

```python
df["Name"] = df["Name"].str.strip().str.title()
df["Sex"] = df["Sex"].str.lower()
df["Cabin"] = df["Cabin"].str.strip()
df["Ticket"] = df["Ticket"].str.strip()
```

```python
df.info()
df.isnull().sum()
df.to_csv("titanic_cleaned.csv", index=False)
df.dtypes
```
