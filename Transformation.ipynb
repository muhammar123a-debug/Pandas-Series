1 Applying functions (.apply(), .map(), .applymap())

```python
import pandas as pd
```

```python
data = pd.read_csv('DataSets/datasets.csv')
```

```python
data.columns = data.columns.str.replace(" ","").str.lower().str.strip()
data.drop_duplicates()
data.drop(columns="clicks", inplace=True)
data.dropna(inplace=True)
data.isna().sum()
clean_active = {
    'Y':1,
    'Yes':1,
    'True':1,
    'N':0,
    'No':0,
    'False':0
}
data['active'] = data["active"].replace(clean_active).astype(int)
data['spend'] = data['spend'].str.replace("$","").astype(float)
data['start_date'] = pd.to_datetime(data['start_date'], errors='coerce')
data['end_date'] = pd.to_datetime(data['end_date'], errors='coerce')
data[data['start_date'] < data['end_date']]
data.loc[data['end_date'] < data['start_date'], 'end_date'] += pd.DateOffset(months=1)
data
```

```python
data['Active_Status'] = data['active'].map({
    1: 'Active',
    0: 'Inactive'
})
data
```

```python
data['channel'].map(lambda x: x.upper())
```

```python
def spend_cat(x):
    if x < 1000:
        return 'Low'
    elif x <= 5000:
        return 'medium'
    else:
        return 'High'
data['spend_cat'] = data['spend'].apply(spend_cat)
```
